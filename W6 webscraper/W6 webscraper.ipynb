```python
# In this project, I build a simple webscraper and store data I scraped as json. 
# I firstly download the target website's HTML using request, 
# then I build a function using BeautifulSoup to get what I want from the website. 
# Finally I store the data as json.

import requests
from bs4 import BeautifulSoup
import pprint
import json
```

# 1. Download target website's HTML

```python
def download_all_htmls():
    htmls = []
    for idx in range(38):
        url = f"http://www.crazyant.net/page/{idx+1}"
        print("craw html:", url)
        r = requests.get(url)
        if r.status_code != 200:
            raise Exception("error")
        htmls.append(r.text)
    return htmls
```

```python
htmls = download_all_htmls()
```

```python
htmls[0]
```

# 2. Analyse HTML

```python
def parse_single_html(html):
    
    soup = BeautifulSoup(html, 'html.parser')
    articles = soup.find_all("article")
    datas = []
    for article in articles:
        title_node = (
            article
            .find("h2", class_="entry-title")
            .find("a")
        )
        title = title_node.get_text()
        link = title_node["href"]
        
        tag_nodes = (
            article
            #.find("footer", class_="entry-footer")
            #.find("span", class_="tags-links")
            .find("a")
        )
        tags = [tag_node.get_text() for tag_node in tag_nodes]
        datas.append(
            {"title":title, "link":link, "tags":tags}
        )
    return datas
```

```python
pprint.pprint(parse_single_html(htmls[0]))
```

```python
all_datas = []
for html in htmls:
    all_datas.extend(parse_single_html(html))
```

# 3. Save the data

```python
with open("all_article_links.json", "w") as fout:
    for data in all_datas:
        fout.write(json.dumps(data, ensure_ascii=False)+"\n")
```
